```python
# 7. What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?
```

```python
count_m=df['hours-per-week'].min()
count_m
```

```python
count_min = (df['hours-per-week'] == df['hours-per-week'].min()).sum()
count_min
```

```python
min_50k =df[(df['hours-per-week']== count_m ) & (df['salary']==' >50K')].shape[0]
min_50k
```

```python
percentage_min_50k = (min_50k/ count_min)*100
percentage_min_50k
```

```python
# 8. What country has the highest percentage of people that earn >50K and what is that percentage?
```

```python
country_count=df[df['salary']==' >50K']['native_country'].value_counts()
country_count
```

```python
max_sal_count= country_count.max()
max_sal_count
```

```python
sum_country_count=country_count.sum()
sum_country_count
```

```python
per_country_count=(max_sal_count/sum_country_count)*100
per_country_count
```
